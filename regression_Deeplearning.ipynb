```python
import pandas as pd
import numpy as np
```

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

```python
concrete_data.shape
```

```python
concrete_data.describe()
```

```python
concrete_data_columns = concrete_data.columns

predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column
```

```python
predictors.head()
```

```python
target.head()
```

```python
n_cols = predictors.shape[1] # number of predictors
```

A. Build a baseline model 

```python
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
model = regression_model()
```

```python
mean_errors=[]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
for i in range(1,50):
    X_train, X_test, y_train, y_test = train_test_split( predictors,target, test_size=0.3, random_state=4)
    model.fit(X_train, y_train, epochs=50, verbose=2)
    y_predict=model.predict(X_test)
    mean_error=mean_squared_error(y_test, y_predict)
    mean_errors.append(mean_error)
```

Mean_squared_errors


```python
mean_errors
```

```python
import statistics 
```

```python
mean_e=statistics.mean(mean_errors)
std_e=statistics.stdev(mean_errors)
```

```python
from tabulate import tabulate
```

```python
table=[['mean','standard deviation'],[mean_e,std_e]]
```

```python
print(tabulate(table, headers='firstrow'))
```

B. Normalize the data

```python
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```python
n_cols = predictors_norm.shape[1]
```

```python
for i in range(1,50):
    X_train, X_test, y_train, y_test = train_test_split( predictors_norm,target, test_size=0.3, random_state=4)
    model.fit(X_train, y_train, epochs=50, verbose=2)
    y_predict=model.predict(X_test)
    mean_error=mean_squared_error(y_test, y_predict)
    mean_errors.append(mean_error)
```

Mean_square_errors

mean_errors

```python
mean_e_n=statistics.mean(mean_errors)
std_e_n=statistics.stdev(mean_errors)
```

```python
table=[['method','mean','standard deviation'],['data without normalization',mean_e,std_e],['data with normalization',mean_e_n,std_e_n]]
```

```python
print(tabulate(table, headers='firstrow'))
```

C. Increate the number of epochs

```python
for i in range(1,50):
    X_train, X_test, y_train, y_test = train_test_split( predictors_norm,target, test_size=0.3, random_state=4)
    model.fit(X_train, y_train, epochs=100, verbose=2)
    y_predict=model.predict(X_test)
    mean_error=mean_squared_error(y_test, y_predict)
    mean_errors.append(mean_error)
```

```python
mean_errors
```

```python
mean_e_100=statistics.mean(mean_errors)
std_e_100=statistics.stdev(mean_errors)
```

```python
table=[['method','mean','standard deviation'],['data without normalization',mean_e,std_e],['data with normalization',mean_e_n,std_e_n],['data with 100epochs',mean_e_100,std_e_100]]
```

```python
print(tabulate(table, headers='firstrow'))
```

D. Increase the number of hidden layers 

```python
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
model = regression_model()
```

```python
for i in range(1,50):
    X_train, X_test, y_train, y_test = train_test_split( predictors_norm,target, test_size=0.3, random_state=4)
    model.fit(X_train, y_train, epochs=100, verbose=2)
    y_predict=model.predict(X_test)
    mean_error=mean_squared_error(y_test, y_predict)
    mean_errors.append(mean_error)
```

```python
mean_errors
```

```python
mean_e_3hl=statistics.mean(mean_errors)
std_e_3hl=statistics.stdev(mean_errors)
```

```python
table=[['method','mean','standard deviation'],['data without normalization',mean_e,std_e],['data with normalization',mean_e_n,std_e_n],['data with 100epochs',mean_e_100,std_e_100],['model with 3hiddenlayers',mean_e_3hl,std_e_3hl]]
```

```python
print(tabulate(table, headers='firstrow'))
```
